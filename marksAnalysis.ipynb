```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D:\Coding Lab\Data Analytics\Beginner level\Student Marks Analysis\students_dataset.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Update_age'] = data['Age'].fillna(data['Age'].median())
```

```python
data["Update_sci"] = data["Science"].fillna(data["Science"].mean())
data["Update_eng"] = data["English"].fillna(data["English"].mean())
data["Update_his"] = data["History"].fillna(data["History"].mean())
data["Update_math"] = data["Maths"].fillna(data["Maths"].mean())
```

```python
data.describe()
```

```python
data["Total"] = round(data["Update_eng"]+data["Update_his"]+data["Update_math"]+data["Update_sci"],2)
data["Average"] = round(data["Total"]/4,2)
```

```python
data.head()
```

```python
def grade(average):
    if average >= 80:
        return "A"
    elif average >=70:
        return "B"
    elif average >=60:
        return "C"
    elif average >=50:
        return "D"
    else:
        return "F"
    
```

```python
data["Grade"] = data["Average"].apply(grade)
data.head()
```

Subject Wise Average

```python
sci_avg = float(data["Science"].sum()/100)
eng_avg = float(data["English"].sum()/100)
his_avg = float(data["History"].sum()/100)
math_avg = float(data["Maths"].sum()/100)
print(f"The Average Marks Of Science Is {sci_avg}.")
print(f"The Average Marks Of English Is {eng_avg}.")
print(f"The Average Marks Of History Is {his_avg}.")
print(f"The Average Marks Of Maths Is {math_avg}.")
```
